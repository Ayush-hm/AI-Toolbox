**Time Resampling**

```python
df=pd.read_csv('/content/starbucks.csv',index_col='Date',parse_dates=True)
df.head()
```

```python
df.info()
```

```python
df.index
```

```python
df.resample(rule='YE').std()
```

```python
def first_day(entry):
  if len(entry):
    return entry[0]
```

```python
df.resample(rule='YE').apply(first_day)
```

```python
df['Close'].resample('M').mean().plot.bar(title='Yearly Close Price')
```

**Rolling & Expanding a Time Series**

```python
df['Close'].plot(figsize=(5,4))
```

```python
df_roll=df.rolling(window=7).mean()
# first n-1(n is the window) values will always be NaN because there are not enough values to take the mean of.
```

```python
df['Close'].plot(figsize=(20,5))
df_roll['Close'].plot()
```

```python
df['Close: 30 Day Mean']=df['Close'].rolling(window=30).mean()
df.head()
```

```python
df[['Close','Close: 30 Day Mean']].plot(figsize=(20,5))
```

```python
df['Close'].expanding().mean().plot()
```
