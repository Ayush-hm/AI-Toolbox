```python
dfd = pd.get_dummies(df.carb, prefix='carb')
new_df = pd.concat([new_df,dfd], axis=1)
del new_df['carb']
print(new_df.head())
# print(new_df.info())
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
mms = new_df[['mpg', 'disp','hp','drat','wt','qsec']]
mmsc=MinMaxScaler(feature_range=(0,1))
d1 = mmsc.fit(mms)
d2=d1.transform(mms)
data_frame = pd.DataFrame(np.round(d2,0)) 
```

```python
data_frame.columns = ['mpg', 'disp','hp','drat','wt','qsec']
data_frame.head()
```

```python
new_df.drop(['mpg', 'disp','hp','drat','wt','qsec'],axis=1,inplace=True)
```

```python
new_df=pd.concat([new_df,data_frame],axis=1)
```
